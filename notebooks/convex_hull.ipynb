```python
import numpy as np
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
rand_vertices = 10 * np.random.random_sample((100, 2)) - 5
rand_vertices
```

```python
hull = ConvexHull(rand_vertices)

plt.plot(rand_vertices[:,0], rand_vertices[:,1], 'o')
for simplex in hull.simplices:
    plt.plot(rand_vertices[simplex, 0], rand_vertices[simplex, 1], 'k-')
plt.plot(rand_vertices[hull.vertices,0], rand_vertices[hull.vertices,1], 'r--', lw=2)
plt.plot(rand_vertices[hull.vertices[0],0], rand_vertices[hull.vertices[0],1], 'ro')
plt.show()
```
